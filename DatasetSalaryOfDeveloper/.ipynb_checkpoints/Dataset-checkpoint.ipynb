### Maaşların ortalması

```python
#Bunun asan yolu var .Çünki ,salary_in_usd column da bütün maaşlar USD -ə çevrilib
data['salary_in_usd'].mean()
```

```python
#en cox en az maas 
print(data['salary_in_usd'].min())
data['salary_in_usd'].max()
```

```python
#en cox maas verilen job
data['salary_in_usd'].max()#job_title
data_row=data[data['salary_in_usd']==data['salary_in_usd'].max()].index
```

```python
data.iloc[data_row]['job_title'][data_row[0]]
```

### GroupBy

```python
print(data.groupby('work_year')['salary_in_usd'])
df=pd.DataFrame(data[['work_year','salary_in_usd']].groupby('work_year')['salary_in_usd'])
df
```

```python
#csv file yaratmaq
df.to_csv("work_salary",index=False)
```

### Pivot table

```python
df_pivot_slary_year=pd.pivot_table(data,values='salary_in_usd',index='work_year',columns='job_title',aggfunc='max')
print(df_pivot_slary_year.loc[2021,'AI Scientist'])
df_pivot_slary_year
```
